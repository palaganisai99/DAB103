```python
# forest Cover

fcover05 = pd.read_csv('fcover0105.csv')
fcover05 = fcover05.drop([0,1])
fcover05 = fcover05.drop(columns = ["Column5","Column9","Column13"])
fcover05.rename(columns=fcover05.iloc[0], inplace = True)
fcover05 = fcover05.drop([2,3])
fcover05 = fcover05.drop([39,40])
fcover05 = fcover05.rename(columns = {"States/UTs":"State/UT","Geographical area":"Geographical_Area","Forest cover in 2001":"2001","Forest cover in 2003":"2003","Forest cover in 2005":"2005"})
fcover05 = fcover05.melt(id_vars=['State/UT','Geographical_Area'],var_name = 'Year',value_name='Forest_cover')
fcover13 = pd.read_csv('fcover13.csv')
fcover13 = fcover13.drop([0,1])
fcover13 = fcover13.drop(columns = ["Column6","Column10","Column14"])
fcover13.rename(columns=fcover13.iloc[0], inplace = True)
fcover13 = fcover13.drop([2,3])
fcover13 = fcover13.drop([39,40])
fcover13 = fcover13.rename(columns = {"States/UTs":"State/UT","Geographical area":"Geographical_Area","Forest cover in 2007*":"2007","Forest cover in 2011":"2011","Forest cover in 2013":"2013"})
fcover13 = fcover13.melt(id_vars=['State/UT','Geographical_Area'],var_name = 'Year',value_name='Forest_cover')
fcover15 = pd.read_csv('fcover15.csv')
fcover15 = fcover15.drop([0,1,39])
fcover15.rename(columns=fcover15.iloc[0], inplace = True)
fcover15 = fcover15.drop([2])
fcover15 = fcover15.drop(columns = ["Total","Percent of geographical area","Scrub"])
fcover15 = fcover15.melt(id_vars=['States/Uts','Geographical area'],var_name = 'Types_of_forest',value_name='Forest_cover')
fcover15 = fcover15.rename(columns = {"States/Uts":"State/UT","Geographical area":"Geographical_Area"})
fcover15["Year"] = "2015"
colnames = ["State/UT","Geographical_Area","Year","Forest_cover","Types_of_forest"]
fcover15 = fcover15.reindex(columns = colnames)
forest_cover = pd.DataFrame()
forest_cover = forest_cover.append([fcover05,fcover13,fcover15])
forest_cover['Geographical_Area'] = forest_cover['Geographical_Area'].str.replace(',','')
forest_cover['Forest_cover'] = forest_cover['Forest_cover'].str.replace(',','')

c = forest_cover.iloc[:,3]
plt.hist(c, bins='auto')
plt.title("Histogram of Forest Cover")
plt.xlabel("Forest Cover in Km2")
```

```python
# Mangrove Cover
mangrove_cover = pd.read_csv("mangrove.csv")
mangrove_cover = mangrove_cover.drop([0,1])
mangrove_cover.rename(columns=mangrove_cover.iloc[0], inplace = True)

mangrove_cover = mangrove_cover.drop([2,15])
mangrove_cover = mangrove_cover.melt(id_vars=['State/UT'],var_name = 'Year',value_name='Mangrove_cover')
mangrove_cover['Mangrove_cover'] = mangrove_cover['Mangrove_cover'].str.replace(',','')
print(mangrove_cover)

c = mangrove_cover.iloc[:,2]
plt.hist(c, bins='auto')
plt.title("Histogram of Mangrove Cover")
plt.xlabel("Mangrove Cover in Km2")
```
